# Read multiple CSV files

```python
# Get all CSV files in the folder
all_files = glob(file_path)
```

```python
df = pd.concat(
    (pd.read_csv(f, index_col="Date/Time", parse_dates=True) for f in all_files)
)
df.head()
```

```python
# Drop columns not needed

df.dropna(axis=1, how="all", inplace=True)  # Columns with all NA
df.drop(
    df.columns[df.columns.str.contains("Flag")], axis=1, inplace=True
)  # Columns with "Flag"
df.drop(
    ["Longitude (x)", "Latitude (y)", "Snow on Grnd (cm)", "Cool Deg Days (°C)"],
    axis=1,
    inplace=True,
)  # Wont be used
df.head()
```

```python
df = df.rename(
    columns={
        "Station Name": "Station",
        "Climate ID": "ID",
        "Max Temp (°C)": "Max_T",
        "Min Temp (°C)": "Min_T",
        "Mean Temp (°C)": "Mean_T",
        "Heat Deg Days (°C)": "HDD",
        "Total Precip (mm)": "Precip_T",
    }
)
```

```python
# Info

df.info()
```

```python
df.to_pickle(save_path + "climate_df.pkl")
```
